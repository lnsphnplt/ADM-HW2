In our data set location_id percentages are high and we should try and fill in these gaps if we want to work with these values. We defined the function below to fill in na values if needed. It makes our data set more reliable and better to work with for tasks like machine learning.

```python

# replacing na with zero values
data_profiles.following.fillna(0, inplace = True)
data_profiles.followers.fillna(0, inplace = True)
data_profiles.n_posts.fillna(0, inplace = True)
data_profiles.profile_id.fillna(0, inplace = True)
data_profiles.firstname_lastname.fillna('', inplace = True)
data_profiles.description.fillna('', inplace = True)
data_profiles.is_business_account.fillna(False, inplace = True)

# replace special col such as time beforehand to avoid filling with str

def replace_val(inp):
    if inp == 'int64':
        return 0
    elif inp == 'float64':
        return float(0)
    elif inp == 'bool':
        return 'False'
    elif inp == 'str':
        return '' 

def replace_na(data_frame):
    for col in data_frame:
        data = data_frame[col]
        dtype = data.dtype
        if dtype == 'object':
            sample = np.where(data != data.isna())
            sample = data[sample[0][2]]
            if sample == False or sample == True:
                dtype = 'bool'
            else:
                dtype = 'str'
        replace = replace_val(dtype)
        data.fillna(replace, inplace = True)

    return data_frame
```

```python
# adding a date_time column
data_profiles['date_time']= pd.to_datetime(data_profiles.cts)
```

```python
# function takes time_segments, data_profile(profiles.csv in pd format)
# (# requires profile data with cts in format to_datetime column)
# time_seq must be circular including both max==min limits

def posts_per_timeinterval(time_seq, data_profile): 
    sub_times = data_profile.dropna(subset = ['cts']) # drops profiles with cts/data_time NA.
    times = sub_times.date_time
    hours = np.array([time.hour for time in times])
    hist, bins = np.histogram(hours, time_seq)
    bins = bins[:-1]
    w = [int(time_seq[i]-time_seq[i-1])for i, time in enumerate(time_seq)][1:]
    
    fig, ax = plt.subplots()
    ax.bar(bins, hist, width = w, align = 'edge', color=['blue', 'green'])
    ax.set_xlim(0, 24)

    return ax
```

```python
# function that takes only the cts column, converts it to date_time
# necessary when handling larger files (such as posts.csv)

def posts_per_timeinterval_only_time_col(time_seq, time_col): 
    sub_times = time_col.dropna() # drops profiles with cts/data_time NA.
    times = pd.to_datetime(sub_times)
    hours = np.array([time.hour for time in times])
    hist, bins = np.histogram(hours, time_seq)
    bins = bins[:-1]
    w = [int(time_seq[i]-time_seq[i-1])for i, time in enumerate(time_seq)][1:]
    
    fig, ax = plt.subplots()
    ax.bar(bins, hist, width = w, align = 'edge', color=['blue', 'green'])
    ax.set_xlim(0, 24)

    return ax
```

```python
# function returns the post_ids found in posts.csv for a given profile_id
def postids_of_id (profile_id, posts):
    
    col_list = ['post_id', 'profile_id']
    post_idss = []
    post_idxs = []
    
    i=0
    for chunk in pd.read_csv(posts+'', sep = '\t', usecols = col_list, header = 0, chunksize = 100000):
        chunk.dropna(subset = ['profile_id'])
        profile_ids = np.array(chunk.profile_id)
        idx = np.where(profile_ids == profile_id )
        if idx[0].size >0:

            post_ids = np.array(chunk.post_id)[idx]
            post_idss.append(post_ids)
            post_idxs.append(idx[0]+100000*i)
        i+=1
    post_idss = np.concatenate(post_idss)
    post_idxs = np.concatenate(post_idxs)
    
    return post_idss, post_idxs
```

```python
# function return rel post information from posts.csv for given profile_id
# all in one more efficient ?

def posts_from_id (profile_id, posts):
    
    col_list = ['post_id','profile_id', 'location_id', 'cts','post_type',  'numbr_likes', 'number_comments']
    post_list = []
    
    i=0
    for chunk in pd.read_csv(posts+'', sep = '\t', usecols = col_list, header = 0, chunksize = 100000):
        #print(i)
        chunk.dropna(subset = ['profile_id'])
        profile_ids = np.array(chunk.profile_id)
        idx = np.where(profile_ids == profile_id )
        if idx[0].size >0:
            inter = chunk[['location_id', 'cts','post_type',  'numbr_likes', 'number_comments']].iloc[idx]
            post_list.append(inter)
        #i+=1
    a = pd.concat(post_list)
    
    
    return a
```

```python
# top n posting take 2
def top_nr_posts(n, posts):
    
    col_list = ['profile_id']
    post_list = []
    i =0
    for chunk in pd.read_csv(posts+'', sep = '\t', usecols = col_list, header = 0, chunksize = 100000):
        #print(i)
        chunk.dropna(subset = ['profile_id'])
        sizes = chunk.groupby('profile_id').size().sort_values(ascending=False)

        sizes = sizes.iloc[:n]
        post_list.append(sizes)

    #i+=1

    
    return post_list
```

```python
# function takes n = top n posting, posts.csv df
# passes scanning to function top_nr_posts which returns fragments of top n posting accounts

# bits of top n are concatenated and new df formed, sorted and overall top n selected
# returns list of top posting profile_ids

def get_top_nr_posts(n, posts):
    post_list = top_nr_posts(n, posts)
    conc_df = pd.concat(post_list)
    
    zipped = list(zip(conc_df.index, conc_df.values))[:-1]
    summ = pd.DataFrame(zipped, columns=['poster_id', 'nr_count']).groupby('poster_id')['nr_count'].sum()#.sort_values('nr_count',ascending=False)
    zipped = list(zip(summ.index, summ.values))
    summed = pd.DataFrame(zipped, columns=['poster_id', 'nr_count']).sort_values('nr_count', ascending = False)[:n]
    
    
    return list(summed.poster_id)
```

### Likelihood that a user return to a location = a Profile tags the same location more than once

```python
# returning to locaitons
# takes posts.csv, groups by location_id and makes df with [location_id, list(profile_ids that tagged that location)]
def profile_locs(posts):
    
    col_list = ['profile_id', 'location_id']
    all_list = []
    i = 0
    once = 0
    more = 1
    vis_once= {}
    locs= {}
    
    for chunk in pd.read_csv(posts+'', sep = '\t', usecols = col_list, header = 0, chunksize = 100000):
        chunk = chunk.dropna(subset = ['location_id']).dropna(subset = ['profile_id'])
        grouped = chunk.groupby(by = 'location_id')['profile_id'].agg(lambda
                             x: list(x))#.apply(list)
        ind = np.array(grouped.index)
        
        sub_frame = pd.DataFrame(grouped.values, ind)
        all_list.append(sub_frame)
        
    return all_list
```

```python
posts = '~/instagram_posts.zip'
loc_ids_profile_ids = profile_locs(posts)
```

```python
# concatenate the df, set column.headers, concatenate lists of same location_id

loc_ids_profile_ids_df = pd.concat(loc_ids_profile_ids)
loc_ids_profile_ids_df.columns = [['location_id'],['profile_id']]

loc_ids_profile_ids_df.reset_index(inplace=True)
loc_ids_profile_ids_df = loc_ids_profile_ids_df#.rename(columns = {'index':'location_id'})
loc_ids_profile_ids_df.columns = ['location_id','profile_id']

grouped_locid = loc_ids_profile_ids_df.groupby(by = 'location_id')['profile_id'].agg(lambda x: list(x))

#(lambda x: np.ndarray.flatten(np.array(list(x))).tolist())
                                                                                     
```

```python
# check how many people tagged a location 1 time vs >1 times
all_tags_profiles = 0
mult_times_tagged = 0

for loc in grouped_locid:

    arr = np.array(loc) #.astype('int32')
    ar = np.ndarray.flatten(arr)

    uniqe, inverse = np.unique(ar, return_inverse=True)
    bincount = np.bincount(inverse)
    
    all_tags_profiles += len(bincount)
    for p in bincount:
        if p>1:
            mult_times_tagged += 1
    
```

```python
print(all_tags_profiles, mult_times_tagged)
ratio = mult_times_tagged/all_tags_profiles
print('Ratio of places that were tagged multiple times by the same person to just once: ' + str(round(ratio, 3)) )
      
```

Looking at all posts and dropping those for which location_id or profile_id are not available we list all profiles that have tagged a location. If a profile appears more than once for a location_id, this indicates that the person has returned to the place more than once.
We interprete the probability that a user returns to a place as the number of locations he/she has tagged more than once / the number of all the different locations he/she tagged. Our calculation results in the **Ratio of places that were tagged multiple times to 0.039.**
This number is quite low, which makes sense if we consider that location_ids are assigned up to a specificity of streets, and the number of different location_ids is therefor enormous. The number doesn't approach 0 because people might tag locations in vicinity to their homes more frequently or post multiple posts from their holiday destination.
